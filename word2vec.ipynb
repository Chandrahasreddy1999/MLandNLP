## Word2vec

```python
import pandas as pd
import re
import nltk
nltk.download("stopwords")
messages=pd.read_csv("SMSSpamCollection",sep="\t",names=["labels","message"])
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
ls=WordNetLemmatizer()
```

```python
## Lemmatizer stemmer
corpus=[]
for message in messages["message"]:
    if isinstance(message,str):
        review=re.sub("[^a-zA-Z]"," ",message)
        review=review.lower()
        review=review.split()
        review=[ls.lemmatize(word,pos="v") for word in review]
        review=" ".join(review)
        corpus.append(review)
```

```python
corpus
```

```python
[[i, j, k] for i, j, k in zip(list(map(len, corpus)), corpus, messages['message']) if i<1] # to find missing values
```

```python
## one more method for data preprocessing
from nltk import sent_tokenize
from gensim.utils import simple_preprocess    
```

```python
words=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
words
```

```python
import gensim
import numpy as np
```

```python
## lets train word2vec from scratch
model=gensim.models.Word2Vec(words)
```

```python
def avg_word2vec(doc):
    # remove out of vocabulary words 
    # sent = [word for word in doc if word in model.wv.index_to_key]
    # print(sent)
    word_vectore=[model.wv[word] for word in doc if word in model.wv.index_to_key]
    return np.mean(word_vectore,axis=0)
```

```python
##!pip install tqdm
```

```python
from tqdm import tqdm
```

```python
# apply for the entire sentences
X=[]
for i in tqdm(range(len(words))):
    X.append(avg_word2vec(words[i]))
```

```python
##Dependent features
y=messages[list(map(lambda x: len(x) >0,corpus))]# to remove missing values
y=pd.get_dummies(y["labels"])
y=y.iloc[:,0].values
```
