```python
import pandas as pd
```

```python
messages=pd.read_csv("SMSSpamCollection",sep="\t",names=["lables","message"])
```

```python
messages.head()
```

```python
import nltk
import re

from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
ps=PorterStemmer()
ss=SnowballStemmer("english")
ws=WordNetLemmatizer()
```

```python
corpus=[]
for message in messages["message"]:
    if isinstance(message,str):
        review=re.sub("[^a-zA-Z]"," ",message)
        review=review.lower()
        review=review.split()
        review=[ws.lemmatize(word) for word in review if not word in stopwords.words("english")]
        review=" ".join(review)
        corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfvc=TfidfVectorizer(max_features=1000,ngram_range=(2,2))
```

```python
X=tfvc.fit_transform(corpus).toarray()
```
